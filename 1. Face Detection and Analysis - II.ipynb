Here in this module we will predict Age, Gender, race and emotion of a human face picture using deepface.
You will need to install the Deepface package using  --pip install deepface--

Theory-
DeepFace is the most lightweight face recognition and facial attribute analysis library for Python. The open-sourced DeepFace library includes all leading-edge AI models for face recognition and automatically handles all procedures for facial recognition in the background.

Set of features available by deepface:
1.Face Verification.
2.Face Recognition.
3.Facial Attribute Analysis. - used in this module
4.Real-Time Face Analysis

The Facial Attribute Analysis is furthur done as follows:
- The input image is firstly recognised for different allignments.
- Face processing of the input image from many to one and one to many.
- Training the image data for feature extraction and testing.
- Applying Euclidean distance and loss function.
- Last step is face matching and categorising into age, gende etc.


For more information on deepface: https://github.com/serengil/deepface

```python
# Read images

img1_path = 'AI/Pictures/emilia1.jpg'
img2_path = 'AI/Pictures/monica2.jpg'
img3_path = 'AI/Pictures/gema.jpg'
img4_path = 'AI/Pictures/emilia2.jpg'



img1 = mpimg.imread(img1_path)
img2 = mpimg.imread(img2_path)
img3 = mpimg.imread(img3_path)
img4 = mpimg.imread(img4_path)

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax1.imshow(img1)
ax1.set_title("emilia")
ax2.imshow(img2)
ax2.set_title("monica")
ax3.imshow(img3)
ax3.set_title("gema")
ax4.imshow(img4)
ax4.set_title("zoe")


print("Pictures present in picture directory")
plt.show()
```

```python
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN

def extact_face_information():
    while True:
        print("\n Select image for more information" "\n Emilia = press 1" " Monica = press 2" " Gema = press 3" " Zoe = press 4")    
        choice = input()
        if choice == '1':
            emilia = img1_path
            img1 = cv2.imread(emilia)        
            obj = DeepFace.analyze(img_path = emilia, actions = ['age', 'gender', 'race', 'emotion'])
            imgplot = plt.imshow(img1)
            plt.show()
        elif choice == '2':
            monica = img2_path
            img2 = cv2.imread(monica)
            obj = DeepFace.analyze(img_path = monica, actions = ['age', 'gender', 'race', 'emotion'])
            imgplot = plt.imshow(img2)
            plt.show()
        elif choice == '3':
            gema = img3_path
            img3 = cv2.imread(gema)
            obj = DeepFace.analyze(img_path = gema, actions = ['age', 'gender', 'race', 'emotion'])
            imgplot = plt.imshow(img3)
            plt.show()
        elif choice == '4':
            zoe = img4_path
            img4 = cv2.imread(zoe)
            obj = DeepFace.analyze(img_path = zoe, actions = ['age', 'gender', 'race', 'emotion'])
            imgplot = plt.imshow(img4)
            plt.show()
        else:
            break
        print("using deepface:", "\n Age:",obj["age"]," Yrs \n","Race:",obj["dominant_race"],"\n","Emotion:", obj["dominant_emotion"],"\n","Gender:", obj["gender"])
    
extact_face_information()
```
